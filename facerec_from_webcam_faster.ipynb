```python
import face_recognition
import cv2
import numpy as np
from mtcnn.mtcnn import MTCNN

# This is a demo of running face recognition on live video from your webcam. It's a little more complicated than the
# other example, but it includes some basic performance tweaks to make things run a lot faster:
#   1. Process each video frame at 1/4 resolution (though still display it at full resolution)
#   2. Only detect faces in every other frame of video.

# PLEASE NOTE: This example requires OpenCV (the `cv2` library) to be installed only to read from your webcam.
# OpenCV is *not* required to use the face_recognition library. It's only required if you want to run this
# specific demo. If you have trouble installing it, try any of the other demos that don't require it instead.

# Get a reference to webcam #0 (the default one)

#video_capture = cv2.VideoCapture(0)
video_capture = cv2.VideoCapture("/home/kevin/reconocimiento/PRUEBA_20190806/PRUEBA.avi")
#video_capture = cv2.VideoCapture("/home/kevin/reconocimiento/facerec_from_webcam_faster/movie_resized.mp4")

########## FOTOS KEVIN

# Load a sample picture and learn how to recognize it.
#kevin_image = face_recognition.load_image_file("kevin.jpg")
#kevin_face_encoding = face_recognition.face_encodings(kevin_image)[0]

########## FOTOS MARIUXI

# Load a sample picture and learn how to recognize it.
mariuxi_image1 = face_recognition.load_image_file("data_trabajo/MARIUXI/ABAJO.jpeg")
mariuxi_face_encoding1 = face_recognition.face_encodings(mariuxi_image1)[0]

mariuxi_image2= face_recognition.load_image_file("data_trabajo/MARIUXI/ABAJO 2.jpeg")
mariuxi_face_encoding2 = face_recognition.face_encodings(mariuxi_image2)[0]

mariuxi_image3 = face_recognition.load_image_file("data_trabajo/MARIUXI/ARRIBA.jpeg")
mariuxi_face_encoding3 = face_recognition.face_encodings(mariuxi_image3)[0]

mariuxi_image4 = face_recognition.load_image_file("data_trabajo/MARIUXI/ARRIBA 2.jpeg")
mariuxi_face_encoding4 = face_recognition.face_encodings(mariuxi_image4)[0]

mariuxi_image5 = face_recognition.load_image_file("data_trabajo/MARIUXI/FRENTE.jpeg")
mariuxi_face_encoding5 = face_recognition.face_encodings(mariuxi_image5)[0]

mariuxi_image6 = face_recognition.load_image_file("data_trabajo/MARIUXI/LADO.jpeg")
mariuxi_face_encoding6 = face_recognition.face_encodings(mariuxi_image6)[0]

mariuxi_image7 = face_recognition.load_image_file("data_trabajo/MARIUXI/LADO 2.jpeg")
mariuxi_face_encoding7 = face_recognition.face_encodings(mariuxi_image7)[0]

mariuxi_image8 = face_recognition.load_image_file("data_trabajo/MARIUXI/LADO 3.jpeg")
mariuxi_face_encoding8 = face_recognition.face_encodings(mariuxi_image8)[0]

mariuxi_image9 = face_recognition.load_image_file("data_trabajo/MARIUXI/LADO 4.jpeg")
mariuxi_face_encoding9 = face_recognition.face_encodings(mariuxi_image9)[0]

mariuxi_image11 = face_recognition.load_image_file("data_trabajo/MARIUXI/PERFIL 2.jpeg")
mariuxi_face_encoding11 = face_recognition.face_encodings(mariuxi_image11)[0]

########## FOTOS JOHANNA

# Load a sample picture and learn how to recognize it.
johanna_image1 = face_recognition.load_image_file("data_trabajo/JOHANNA/ABAJO.jpeg")
johanna_face_encoding1 = face_recognition.face_encodings(johanna_image1)[0]

johanna_image2= face_recognition.load_image_file("data_trabajo/JOHANNA/ABAJO 2.jpeg")
johanna_face_encoding2 = face_recognition.face_encodings(johanna_image2)[0]

johanna_image3 = face_recognition.load_image_file("data_trabajo/JOHANNA/ARRIBA.jpeg")
johanna_face_encoding3 = face_recognition.face_encodings(johanna_image3)[0]

johanna_image4 = face_recognition.load_image_file("data_trabajo/JOHANNA/ARRIBA 2.jpeg")
johanna_face_encoding4 = face_recognition.face_encodings(johanna_image4)[0]

johanna_image5 = face_recognition.load_image_file("data_trabajo/JOHANNA/FRENTE.jpeg")
johanna_face_encoding5 = face_recognition.face_encodings(johanna_image5)[0]

johanna_image6 = face_recognition.load_image_file("data_trabajo/JOHANNA/LADO.jpeg")
johanna_face_encoding6 = face_recognition.face_encodings(johanna_image6)[0]

johanna_image7 = face_recognition.load_image_file("data_trabajo/JOHANNA/LADO 3.jpeg")
johanna_face_encoding7 = face_recognition.face_encodings(johanna_image7)[0]

johanna_image9 = face_recognition.load_image_file("data_trabajo/JOHANNA/PERFIL.jpeg")
johanna_face_encoding9 = face_recognition.face_encodings(johanna_image9)[0]

johanna_image10 = face_recognition.load_image_file("data_trabajo/JOHANNA/PERFIL 2.jpeg")
johanna_face_encoding10 = face_recognition.face_encodings(johanna_image10)[0]

########## FOTOS IVETTE

# Load a sample picture and learn how to recognize it.
ivette_image1 = face_recognition.load_image_file("data_trabajo/IVETTE/1.jpg")
ivette_face_encoding1 = face_recognition.face_encodings(ivette_image1)[0]

ivette_image5 = face_recognition.load_image_file("data_trabajo/IVETTE/5.jpg")
ivette_face_encoding5 = face_recognition.face_encodings(ivette_image5)[0]

ivette_image6 = face_recognition.load_image_file("data_trabajo/IVETTE/6.jpg")
ivette_face_encoding6 = face_recognition.face_encodings(ivette_image6)[0]

ivette_image8 = face_recognition.load_image_file("data_trabajo/IVETTE/8.jpg")
ivette_face_encoding8 = face_recognition.face_encodings(ivette_image8)[0]

ivette_image12 = face_recognition.load_image_file("data_trabajo/IVETTE/11.jpg")
ivette_face_encoding12 = face_recognition.face_encodings(ivette_image12)[0]

########## FOTOS LEONARDO

# Load a sample picture and learn how to recognize it.
leonardo_image1 = face_recognition.load_image_file("data_trabajo/LEONARDO/ABAJO.jpeg")
leonardo_face_encoding1 = face_recognition.face_encodings(leonardo_image1)[0]

leonardo_image2 = face_recognition.load_image_file("data_trabajo/LEONARDO/ARRIBA.jpeg")
leonardo_face_encoding2 = face_recognition.face_encodings(leonardo_image2)[0]

leonardo_image3 = face_recognition.load_image_file("data_trabajo/LEONARDO/ARRIBA 2.jpeg")
leonardo_face_encoding3 = face_recognition.face_encodings(leonardo_image3)[0]

leonardo_image4 = face_recognition.load_image_file("data_trabajo/LEONARDO/FRENTE.jpeg")
leonardo_face_encoding4 = face_recognition.face_encodings(leonardo_image4)[0]

leonardo_image6 = face_recognition.load_image_file("data_trabajo/LEONARDO/LADO 2.jpeg")
leonardo_face_encoding6 = face_recognition.face_encodings(leonardo_image6)[0]

leonardo_image9 = face_recognition.load_image_file("data_trabajo/LEONARDO/LADO 5.jpeg")
leonardo_face_encoding9 = face_recognition.face_encodings(leonardo_image9)[0]

leonardo_image11 = face_recognition.load_image_file("data_trabajo/LEONARDO/âˆ©PERFIL 3.jpeg")
leonardo_face_encoding11 = face_recognition.face_encodings(leonardo_image11)[0]

leonardo_image12 = face_recognition.load_image_file("data_trabajo/LEONARDO/PERFIL 2.jpeg")
leonardo_face_encoding12 = face_recognition.face_encodings(leonardo_image12)[0]

########## FOTOS NANCY

nancy_image1 = face_recognition.load_image_file("data_trabajo/NANCY/1.jpg")
nancy_face_encoding1 = face_recognition.face_encodings(nancy_image1)[0]

nancy_image4 = face_recognition.load_image_file("data_trabajo/NANCY/4.jpg")
nancy_face_encoding4 = face_recognition.face_encodings(nancy_image4)[0]

nancy_image5 = face_recognition.load_image_file("data_trabajo/NANCY/5.JPG")
nancy_face_encoding5 = face_recognition.face_encodings(nancy_image5)[0]

nancy_image7 = face_recognition.load_image_file("data_trabajo/NANCY/7.JPG")
nancy_face_encoding7 = face_recognition.face_encodings(nancy_image7)[0]

nancy_image8 = face_recognition.load_image_file("data_trabajo/NANCY/8.JPG")
nancy_face_encoding8 = face_recognition.face_encodings(nancy_image8)[0]

nancy_image9 = face_recognition.load_image_file("data_trabajo/NANCY/9.JPG")
nancy_face_encoding9 = face_recognition.face_encodings(nancy_image9)[0]

nancy_image10 = face_recognition.load_image_file("data_trabajo/NANCY/10.JPG")
nancy_face_encoding10 = face_recognition.face_encodings(nancy_image10)[0]

nancy_image11 = face_recognition.load_image_file("data_trabajo/NANCY/11.JPG")
nancy_face_encoding11 = face_recognition.face_encodings(nancy_image11)[0]

########## FOTOS VICTOR

victor_image1 = face_recognition.load_image_file("data_trabajo/VICTOR/1.JPG")
victor_face_encoding1 = face_recognition.face_encodings(victor_image1)[0]

victor_image3 = face_recognition.load_image_file("data_trabajo/VICTOR/3.JPG")
victor_face_encoding3 = face_recognition.face_encodings(victor_image3)[0]

victor_image4 = face_recognition.load_image_file("data_trabajo/VICTOR/4.JPG")
victor_face_encoding4 = face_recognition.face_encodings(victor_image4)[0]

victor_image6 = face_recognition.load_image_file("data_trabajo/VICTOR/6.JPG")
victor_face_encoding6 = face_recognition.face_encodings(victor_image6)[0]

victor_image7 = face_recognition.load_image_file("data_trabajo/VICTOR/7.JPG")
victor_face_encoding7 = face_recognition.face_encodings(victor_image7)[0]

victor_image11 = face_recognition.load_image_file("data_trabajo/VICTOR/11.JPG")
victor_face_encoding11 = face_recognition.face_encodings(victor_image11)[0]

victor_image13 = face_recognition.load_image_file("data_trabajo/VICTOR/13.JPG")
victor_face_encoding13 = face_recognition.face_encodings(victor_image13)[0]

# Create arrays of known face encodings and their names
known_face_encodings = [
    #kevin_face_encoding,
    mariuxi_face_encoding1,
    mariuxi_face_encoding2,
    mariuxi_face_encoding3,
    mariuxi_face_encoding4,
    mariuxi_face_encoding5,
    mariuxi_face_encoding6,
    mariuxi_face_encoding7,
    mariuxi_face_encoding8,
    mariuxi_face_encoding9,
    mariuxi_face_encoding11,
    johanna_face_encoding1,
    johanna_face_encoding2,
    johanna_face_encoding3,
    johanna_face_encoding4,
    johanna_face_encoding5,
    johanna_face_encoding6,
    johanna_face_encoding7,
    johanna_face_encoding9,
    johanna_face_encoding10,
    ivette_face_encoding1,
    ivette_face_encoding5,
    ivette_face_encoding6,
    ivette_face_encoding8,
    ivette_face_encoding12,
    leonardo_face_encoding1,
    leonardo_face_encoding2,
    leonardo_face_encoding3,
    leonardo_face_encoding4,
    leonardo_face_encoding6,
    leonardo_face_encoding9,
    leonardo_face_encoding11,
    leonardo_face_encoding12,
    nancy_face_encoding1,
    nancy_face_encoding4,
    nancy_face_encoding5,
    nancy_face_encoding7,
    nancy_face_encoding8,
    nancy_face_encoding9,
    nancy_face_encoding10,
    nancy_face_encoding11,
    victor_face_encoding1,
    victor_face_encoding3,
    victor_face_encoding4,
    victor_face_encoding6,
    victor_face_encoding7,
    victor_face_encoding11,
    victor_face_encoding13
]
known_face_names = [
    #"Kevin",
    "Mariuxi",
    "Mariuxi",
    "Mariuxi",
    "Mariuxi",
    "Mariuxi",
    "Mariuxi",
    "Mariuxi",
    "Mariuxi",
    "Mariuxi",
    "Mariuxi",
    "Johanna",
    "Johanna",
    "Johanna",
    "Johanna",
    "Johanna",
    "Johanna",
    "Johanna",
    "Johanna",
    "Johanna",
    "Ivette",
    "Ivette",
    "Ivette",
    "Ivette",
    "Ivette",
    "Leonardo",
    "Leonardo",
    "Leonardo",
    "Leonardo",
    "Leonardo",
    "Leonardo",
    "Leonardo",
    "Leonardo",
    "Nancy",
    "Nancy",
    "Nancy",
    "Nancy",
    "Nancy",
    "Nancy",
    "Nancy",
    "Nancy",
    "Victor",
    "Victor",
    "Victor",
    "Victor",
    "Victor",
    "Victor",
    "Victor"
    
]

# Initialize some variables
face_locations = []
face_encodings = []
face_names = []
process_this_frame = True
detector = MTCNN()

while True:
    # Grab a single frame of video
    ret, frame = video_capture.read()

    # Resize frame of video to 1/4 size for faster face recognition processing
    small_frame = cv2.resize(frame, (0, 0), fx=1, fy=1)

    # Convert the image from BGR color (which OpenCV uses) to RGB color (which face_recognition uses)
    rgb_small_frame = small_frame[:, :, ::-1]
    data = detector.detect_faces(rgb_small_frame)
    
    # Only process every other frame of video to save time
    if process_this_frame and data:
        
        bounding_box = data[0]['box']
        x = bounding_box[0]
        y = bounding_box[1]
        w = bounding_box[0] + bounding_box[2]
        h = bounding_box[1] + bounding_box[3]
        face_locations= [(y, w, h, x)]
        #face_locations= [(0, w-x, h-y, 0)]
        crop_imag = rgb_small_frame[y:h, x:w]
                
        # Find all the faces and face encodings in the current frame of video
        #face_locations = face_recognition.face_locations(rgb_small_frame)
                
        face_encodings = face_recognition.face_encodings(crop_imag, [(0, w-x, h-y, 0)]) #[] este corchete es face locations
        #face_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations)
        
        cv2.imshow("crop",crop_imag)
        
        face_names = []
        for face_encoding in face_encodings:
            # See if the face is a match for the known face(s)
            matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
            name = "Unknown"

            # # If a match was found in known_face_encodings, just use the first one.
            # if True in matches:
            #     first_match_index = matches.index(True)
            #     name = known_face_names[first_match_index]

            # Or instead, use the known face with the smallest distance to the new face
            face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
            best_match_index = np.argmin(face_distances)
            if matches[best_match_index]:
                name = known_face_names[best_match_index]
                
            face_names.append(name)

    process_this_frame = not process_this_frame


    # Display the results
    for (top, right, bottom, left), name in zip(face_locations, face_names):
        # Scale back up face locations since the frame we detected in was scaled to 1/4 size
        top *= 1
        right *= 1
        bottom *= 1
        left *= 1

        # Draw a box around the face
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 1)

        # Draw a label with a name below the face
        #cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left + 6, bottom - 6), font, 0.5, (255, 255, 255), 1)

    # Display the resulting image
    cv2.imshow('Video', frame)

    # Hit 'q' on the keyboard to quit!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release handle to the webcam
video_capture.release()
cv2.destroyAllWindows()
```
